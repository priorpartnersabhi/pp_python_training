```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#geopandas to process geospatial/gis data
import geopandas as gpd
```

```python
#User input
ethnicity_csv = "N:\Geodatabase\Raw_Data\Census 2021\Ethnicity\TS021-2021-3-filtered-2025-02-06T08_07_15Z.csv"
```

```python
#Load csv
ethnicity_df = pd.read_csv(ethnicity_csv, nrows= 1000)
ethnicity_df.head(25)
```

```python
lsoa_ethincity_df_P = pd.pivot_table(ethnicity_df, values='Observation', index=['Lower layer Super Output Areas Code'], columns=['Ethnic group (20 categories)'], aggfunc=np.sum)
lsoa_ethincity_df_P = lsoa_ethincity_df_P.reset_index()
lsoa_ethincity_df_P = lsoa_ethincity_df_P.rename_axis("ID")  # Rename index
lsoa_ethincity_df_P.head()
```

```python
# Dictionary for renaming columns with corrected keys
column_rename_map = {
    "Lower layer Super Output Areas Code":"lsoa21cd",
    "Asian, Asian British or Asian Welsh: Bangladeshi": "asian_bangladeshi_count",
    "Asian, Asian British or Asian Welsh: Chinese": "asian_chinese_count",
    "Asian, Asian British or Asian Welsh: Indian": "asian_indian_count",
    "Asian, Asian British or Asian Welsh: Pakistani": "asian_pakistani_count",
    "Asian, Asian British or Asian Welsh: Other Asian": "asian_other_count",
    
    "Black, Black British, Black Welsh, Caribbean or African: African": "black_african_count",
    "Black, Black British, Black Welsh, Caribbean or African: Caribbean": "black_caribbean_count",
    "Black, Black British, Black Welsh, Caribbean or African: Other Black": "black_other_count",
    
    "Mixed or Multiple ethnic groups: Other Mixed or Multiple ethnic groups": "mixed_or_multiple_other_count",
    "Mixed or Multiple ethnic groups: White and Asian": "mixed_or_multiple_white_and_asian_count",
    "Mixed or Multiple ethnic groups: White and Black African": "mixed_or_multiple_white_and_black_african_count",
    "Mixed or Multiple ethnic groups: White and Black Caribbean": "mixed_or_multiple_white_and_black_caribbean_count",
    
    "Other ethnic group: Any other ethnic group": "other_count",
    "Other ethnic group: Arab": "other_arab_count",
    
    "White: English, Welsh, Scottish, Northern Irish or British": "white_british_count",
    "White: Gypsy or Irish Traveller": "white_gypsy_or_irish_count",
    "White: Irish": "white_irish_count",
    "White: Other White": "white_other_count",
    "White: Roma": "white_roma_count"
}

# Rename columns using the dictionary
lsoa_ethincity_df_P.rename(columns=column_rename_map, inplace=True)
lsoa_ethincity_df_P.drop(['Does not apply'],1, inplace=True)
lsoa_ethincity_df_P.head()
```

```python
lsoa_ethincity_df_P.columns
```

```python
lsoa_ethincity_df_P['total_ethnicity_pop'] = lsoa_ethincity_df_P.sum(axis=1,numeric_only=True)
lsoa_ethincity_df_P.head()
```

```python
def createperc(dataframe, columns, total_column):
    for column in columns:
        if pd.api.types.is_numeric_dtype(dataframe[column]):
            perc_column = column.replace('_count', '_perc')          
            dataframe[perc_column] = (dataframe[column]/dataframe[total_column])*100         
```

```python
createperc(lsoa_ethincity_df_P, lsoa_ethincity_df_P.columns,'total_ethnicity_pop')
```
